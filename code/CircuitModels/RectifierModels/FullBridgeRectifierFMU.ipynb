```python
def run_FMU(L, C):
    """
    Runs the Modelica model code.CircuitModels.RectifierModels.FullBridgeRectifier
    Takes L and C as inputs for the inductance and capacitance of the rectifier
    """
    mmodel = load_fmu('/home/sswaminathan/olin_share/ModSim/ModSimPy/code/' + \
                      'CircuitModels/RectifierModels/./code_CircuitModels_RectifierModels_FullBridgeRectifier.fmu')
    mmodel.set('inductor.L', L)
    mmodel.set('capacitor.C', C)
    return mmodel.simulate(final_time = 10)
```

```python
help(run_FMU)
```

```python
res = run_FMU(0.03, 0.022)
```

```python
v_c = pd.Series(res['capacitor.v'], index=res['time'])
v_c.plot()
```

```python
l_range = np.linspace(0.01, 0.05, 5)
c_range = np.linspace(0.01, 0.05, 11)
dat = pd.DataFrame(columns=c_range, index=l_range)
dat[c_range[1]][l_range[1]] = pd.Series(np.linspace(1,10,10), index=np.linspace(1,10,10))
dat[c_range[1]][l_range[1]]
```

```python
def fmu_sweep(L_range, C_range):
    results = pd.DataFrame(index = l_range, columns = c_range)
    for l in L_range:
        for c in C_range:
            output = run_FMU(l, c)
            results[c][l] = pd.Series(output['capacitor.v'], index=output['time'])
    return results
```

```python
res = fmu_sweep(l_range, c_range);
```
